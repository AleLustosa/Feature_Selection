```python
classe1_m = df[(df['Pclass'] == 1) & (df['Sex'] == 'male')]
classe1_f = df[(df['Pclass'] == 1) & (df['Sex'] == 'female')]
classe2_m = df[(df['Pclass'] == 2) & (df['Sex'] == 'male')]
classe2_f = df[(df['Pclass'] == 2) & (df['Sex'] == 'female')]
classe3_m = df[(df['Pclass'] == 3) & (df['Sex'] == 'male')]
classe3_f = df[(df['Pclass'] == 3) & (df['Sex'] == 'female')]
```

```python
classe1_m['Age'] = classe1_m['Age'].fillna('41.281386')
classe1_f['Age'] = classe1_f['Age'].fillna('34.611765')
classe2_m['Age'] = classe2_m['Age'].fillna('30.740707')
classe2_f['Age'] = classe2_f['Age'].fillna('28.722973')
classe3_m['Age'] = classe3_m['Age'].fillna('26.50758')
classe3_f['Age'] = classe3_f['Age'].fillna('21.750000')
```

```python
df_train = pd.concat(([classe1_m ,classe1_f ,classe2_m, classe2_f, classe3_m, classe3_f ]), axis=0)
```

```python
df_train.head(30)
```

```python
title = ['Mlle','Mrs', 'Mr', 'Miss','Master','Don','Rev','Dr','Mme','Ms','Major','Col','Capt','Countess']

def ExtractTitle(name):
    tit = 'missing'
    for item in title :
        if item in name:
            tit = item
    if tit == 'missing':
        tit = 'Mr'
    return tit
```

```python
df_train["Title"] = df_train.apply(lambda row: ExtractTitle(row["Name"]),axis=1)
```

```python
df_train['Sex'] = df_train['Sex'].map({'female': 0, 'male': 1}).astype(int)
```

```python
Y = df_train['Survived'].values
```

```python
# Normalização Z-score
def encode_numeric_zscore(df, name, mean=None, sd=None):
    if mean is None:
        mean = df[name].mean()

    if sd is None:
        sd = df[name].std()

    df[name] = (df[name] - mean) / sd
```

```python
def encode_text_dummy(df, name):
    dummies = pd.get_dummies(df[name])
    for x in dummies.columns:
        dummy_name = "{}-{}".format(name, x)
        df[dummy_name] = dummies[x]
    df.drop(name, axis=1, inplace=True)
```

```python
X = df_train.drop('Survived', axis=1)
```

```python
X['Age'] = X['Age'].astype('float32')
```

```python
X['SibSp'] = X['SibSp'].astype('float32')
X['Parch'] = X['Parch'].astype('float32')
X['Fare'] = X['Fare'].astype('float32')
```

```python
encode_numeric_zscore(X, 'Age')
```

```python
encode_numeric_zscore(X, 'SibSp')
encode_numeric_zscore(X, 'Parch')
encode_numeric_zscore(X, 'Fare')
```

```python
from keras.callbacks import EarlyStopping
import keras.backend as K
```

```python
X.shape
```

```python
model = Sequential()
```

```python
K.clear_session()
model.add(Dense(units=600, activation='relu', kernel_initializer='normal', input_dim=23))
model.add(Dropout(0.25))
model.add(Dense(1200, activation='relu', kernel_initializer='normal'))
model.add(Dropout(0.25))
model.add(Dense(1200, activation='relu', kernel_initializer='normal'))
model.add(Dense(200, activation='relu', kernel_initializer='normal'))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
monitor = EarlyStopping(monitor = 'val_loss', min_delta = 1e-11, patience = 13, verbose = 1, mode = 'auto')
```

```python
model.fit(X_train, y_train, batch_size=25, epochs=700, verbose=1, validation_data = (X_test, y_test), 
              callbacks = [monitor], validation_split=0.21)
```

```python
model.evaluate(X_test, y_test)
```
